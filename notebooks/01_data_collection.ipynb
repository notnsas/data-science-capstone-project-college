## **2. Scrape Reviews**

Bagian ini berisi kode untuk mengambil ulasan aplikasi dari Google Play Store menggunakan library `google-play-scraper`.

Kode berikut mengambil ulasan menggunakan fungsi `reviews`:
- `app_id = 'com.whatsapp'`: Mendefinisikan ID paket aplikasi WhatsApp yang ulasannya akan diambil.
- `scraped_reviews, _ = reviews(...)`: Memanggil fungsi `reviews` untuk mengambil ulasan. Fungsi ini mengembalikan dua nilai: daftar ulasan (`scraped_reviews`) dan token untuk mengambil halaman berikutnya (yang diabaikan dengan `_`).
- `lang='id'`: Menetapkan bahasa ulasan yang diambil ke Bahasa Indonesia.
- `country='id'`: Menetapkan negara asal ulasan yang diambil ke Indonesia.
- `sort=Sort.NEWEST`: Mengurutkan ulasan berdasarkan yang terbaru. `Sort.NEWEST` berasal dari enum `Sort` yang diimpor.
- `count=10000`: Menetapkan jumlah maksimum ulasan yang akan diambil sebanyak 10000.

```python
app_id = 'com.spotify.music'
scraped_reviews, _ = reviews(
    app_id,
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # Use Sort.NEWEST instead of 'newest'
    count=30000 # defaults to 100
)
```

## **4. Convert the data to CSV**

Konversi data ulasan yang diambil menjadi DataFrame pandas dan simpan ke file CSV.

**Reasoning**:
Convert the list of dictionaries containing the scraped reviews into a pandas DataFrame for easier manipulation and then save it as a CSV file.

```python
reviews_df = pd.DataFrame(scraped_reviews)

# Save the DataFrame to a CSV file
reviews_df.to_csv('spotify_reviews_eng_1.csv', index=False)
```
